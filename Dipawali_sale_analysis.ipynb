```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv("Diwali_Sales_Data.csv",encoding="Latin-1")
# to avoide encoding error use "unicode_escape"
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
# drop unrelated /blank columns 
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.columns
```

```python
df.info()
```

```python
# check for null values 
pd.isnull(df).sum()
```

```python
df.shape
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
# change the datatype
df["Amount"]=df["Amount"].astype("int")
```

```python
# rename cols
df.rename(columns={"Marital_Status":"Shaadi"})
```

### EDA 

```python
df.columns
```

```python
ax= sns.countplot(x="Gender",data=df)
```

```python
ax= sns.countplot(x="Gender",data=df)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sns.barplot(x="Gender",y="Amount",data=sales_gen)
```

```python
# females above graph we can see the most of the buyers are females 
```

### Age

```python
sns.countplot(data=df,x="Age Group",hue="Gender")
```

```python
ax=sns.countplot(data=df,x="Age Group",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total amount vs Age Group
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

### State

```python
df.columns
```

```python
sales_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)
```

```python
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(data=sales_state,x="State",y="Orders")
```

```python
# top 10 state 
sales_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(data=sales_state,x="State",y="Orders")
```

```python
# top 10 state 
sales_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(data=sales_state,x="State",y="Amount")
```

```python
# most of oders from up,maharastra, karnataka
```

### Marital Status

```python
ax=sns.countplot(data=df,x="Marital_Status")
sns.set(rc={"figure.figsize":(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(7,5)})
sns.barplot(data=sales_marital,x="Marital_Status",y="Amount",hue="Gender")
```

```python
# married girls shopping more 
```

### Occupation

### Product category

```python
ax=sns.countplot(data=df,x="Product_Category")
sns.set(rc={"figure.figsize":(25,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Product_Category=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
ax=sns.barplot(data=sales_Product_Category,x="Product_Category",y="Amount")
sns.set(rc={"figure.figsize":(25,10)})
```

```python
sales_Product_ID=df.groupby(["Product_ID"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
ax=sns.barplot(data=sales_Product_ID,x="Product_ID",y="Amount")
sns.set(rc={"figure.figsize":(25,10)})
```

```python
sales_Product_ID=df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
ax=sns.barplot(data=sales_Product_ID,x="Product_ID",y="Orders")
sns.set(rc={"figure.figsize":(25,10)})
```
