Categorical data represents variables that have a limited number of unique values or categories. These categories can be unordered or ordered. Examples of categorical data include gender (male/female), color (red/blue/green), or education level (high school/college/graduate). Categorical data is often represented using strings or integers. In Pandas, categorical data is typically stored as a special data type called "Categorical."


Categorical Data Description:

1. value_counts(): Returns the frequency count of each unique category in a categorical column.
2. describe(): Provides summary statistics for categorical data, including count, unique categories, top category, and frequency of the top category.
3. groupby(): Allows you to group and aggregate data based on categorical variables.

Numerical data, on the other hand, represents variables that are continuous or discrete and can take on a range of numerical values. Examples of numerical data include age, height, temperature, or salary. Numerical data is typically represented using numeric data types such as integers or floating-point numbers (decimals). In Pandas, numerical data is commonly stored as "int64" or "float64" data types.

Numerical Data Description:

1. describe(): Generates descriptive statistics for numerical data, including count, mean, standard deviation, minimum, quartiles, and maximum values.
2. mean(), median(), std(), min(), max(): Individual functions to calculate specific statistics for numerical data.
3. groupby(): Enables grouping and aggregation based on categorical variables and performing numerical operations on the grouped data.
